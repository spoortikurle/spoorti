```python
Q1=df['cgpa'].quantile(0.25)
Q3=df['cgpa'].quantile(0.75)
iqr=Q3-Q1
```

```python
upper_limit=Q3+1.5*iqr
lower_limit=Q1-1.5*iqr
```

```python
df[(df['cgpa']>upper_limit)  | (df['cgpa']<lower_limit)]
```

```python
df[(df['cgpa']<upper_limit)&(df['cgpa']>lower_limit)]
```
